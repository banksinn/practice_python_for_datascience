```python
import pandas as pd
df1 = pd.read_csv("SaleOrderPivot.csv")
df1
```

```python
pd.pivot_table(df1,values="ราคา",index="สินค้า") #ไม่มี aggfunc จะได้ mean ออกมาแทน
```

```python
pd.pivot_table(df1,values="ราคา",index=["สินค้า","พนักงานขาย"],aggfunc="sum")
```

```python
pd.pivot_table(df1,values="ราคา",index=["สินค้า","พนักงานขาย"],columns="วันที่ทำรายการ",aggfunc="sum")
```

```python
pd.pivot_table(df1,values="ราคา",index=["สินค้า","พนักงานขาย"],columns="วันที่ทำรายการ",aggfunc="sum",fill_value=0)
```

```python
pd.pivot_table(df1,values="ราคา",index=["สินค้า","พนักงานขาย"],columns="วันที่ทำรายการ",aggfunc="sum",fill_value=0,margins=True)
```

```python
pd.pivot_table(df1,values="ราคา",index=["สินค้า","พนักงานขาย"],columns="วันที่ทำรายการ",aggfunc="sum",fill_value=0,margins=True,margins_name="ยอดรวมทั้งหมด")
```

```python
df_pivot = pd.pivot_table(df1,values="ราคา",index=["สินค้า","พนักงานขาย"],aggfunc="sum",fill_value=0)
df_filter = df_pivot[df_pivot["ราคา"]>2500]
df_filter
```

```python
df2 = pd.read_csv("SaleOrderPivot2.csv")
df2
```

```python
pd.pivot_table(df2,index=["พื้นที่สาขา","สินค้า"]) #ไม่ใส่ values ดังนั้นจะแสดงค่า mean ออกมา
```

```python
pd.pivot_table(df2,values=["จำนวน","ยอดขาย"],index="สินค้า",aggfunc=["sum","min","max"])
```

```python
pd.pivot_table(df2,values=["จำนวน","ยอดขาย"],index="สินค้า",aggfunc={"จำนวน":"sum","ยอดขาย":"mean"})
```
